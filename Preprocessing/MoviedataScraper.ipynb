```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import csv
lijst = []
document = BeautifulSoup(urlopen('http://www.imsdb.com/all%20scripts/').read(), "html5lib")
a = document.find_all('td', {"valign": "top"})[2].find_all('a')
for link in a:
    moviename = link["title"][:-7].replace(":","")
    print("http://www.imsdb.com/scripts/"+moviename.replace(" ", "-")+".html")
    try:
        document2 = BeautifulSoup(urlopen("http://www.imsdb.com/scripts/"+moviename.replace(" ", "-")+".html").read(), "html5lib")
        towrite=' '.join(' '.join([re.sub('<.*?>', '', str(x)) for x in document2.find_all('pre')[-1].contents if "<b" not in str(x)]).split())
        if len(towrite)>20000:
            filename = moviename.replace(" ", "-")+".txt"
            text_file = open("TextFolder/"+filename, "w", encoding="utf-8")
            text_file.write(towrite)
            text_file.close()
            lijst.append([moviename, filename])
        else:
            raise Exception('Document too short')
    except Exception as e:
         lijst.append([moviename, e])
```

```python
import csv
with open('TextFiles.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(lijst)
```

```python
import requests
import json
csvfile = open('TextFiles.csv', 'r')
csvreader = csv.reader(csvfile, delimiter=',')
csvfile2 = open('TextFiles2.csv', 'w', newline='')
csvwriter = csv.writer(csvfile2)
for x in csvreader:
    try:
        url = "http://www.omdbapi.com/?apikey=1b487cac&t="+x[0].replace(" ","+").split(',')[0]
        myResponse = requests.get(url, verify=True)
        if(myResponse.ok):
            jsonname = x[0].replace(" ", "-")+".json"
            if x[1][-4:]== ".txt":
                with open("TextFolder/"+jsonname, 'w') as outfile:
                    result = json.loads(myResponse.text)
                    result["body"] = open("TextFolder/"+x[1], "r").read()
                    json.dump(result, outfile)
                csvwriter.writerow([x[0],x[1],jsonname])
        else:
            csvwriter.writerow([x[0],x[1],myResponse.status_code])
    except Exception as e:
             csvwriter.writerow([x[0],x[1],e])
```
